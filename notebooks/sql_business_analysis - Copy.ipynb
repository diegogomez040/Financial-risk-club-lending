- Understand the relationship between loan characteristics and default rates

- Evaluate the profitability of different loan segments

- Support better credit risk and lending strategy decisions

## Default rates

### Total default rate

```python
%%sql
SELECT 
    COUNT(*) AS total_loans,
    SUM("default") AS total_defaults,
    ROUND(SUM("default")*100.0 / COUNT(*),2) AS default_rate_percent
FROM loans;
```

The overall default rate for the Lending Club portfolio is **21.47%** (294,415 defaults out of 1,371,165 loans). This relatively high level of credit risk highlights the importance of robust underwriting and risk-based pricing strategies. Lenders should carefully monitor segments that contribute most to defaults and adjust approval criteria, interest rates, or credit policies to improve portfolio performance and reduce losses.

### Default rate by credit grade

```python
%%sql
SELECT 
    grade,
    COUNT(*) AS loans,
    SUM("default") AS defaults,
    ROUND(SUM("default")*100.0 / COUNT(*),2) AS default_rate
FROM loans
GROUP BY grade
ORDER BY grade;
```

Default rates increase sharply as credit quality worsens: **Grade A** loans have a default rate of only **6.7%**, while **Grade G** loans default at **51.4%**. Mid-grade segments like **C** and **D** already show elevated risk (**24.23%** and **32.32%** respectively), suggesting that taking on lower-grade borrowers requires substantially higher pricing or tighter approval criteria to compensate for the much higher probability of loss.

### Default rate by income level

```python
%%sql
SELECT 
    CASE
        WHEN annual_inc < 40000 THEN 'Low Income'
        WHEN annual_inc < 80000 THEN 'Middle Income'
        ELSE 'High Income'
    END AS income_group,
    
    COUNT(*) AS loans,
    SUM("default") AS defaults,
    
    ROUND(SUM("default")*100.0 / COUNT(*),2) AS default_rate
    
FROM loans
GROUP BY income_group;
```

Borrowers in the **Low Income** segment have the highest default rate at **25.26%**, compared with **22.39%** for **Middle Income** and **18.49%** for **High Income** borrowers. This pattern suggests that income capacity is a key driver of repayment ability, and that loans to lower-income borrowers should be monitored more closely, potentially with stricter affordability checks or smaller ticket sizes to mitigate risk.

### Default rate by loan amount

```python
%%sql
SELECT 
    CASE
        WHEN loan_amnt < 5000 THEN 'Small'
        WHEN loan_amnt < 15000 THEN 'Medium'
        ELSE 'Large'
    END AS loan_size,
    
    COUNT(*) AS loans,
    SUM("default") AS defaults,
    
    ROUND(SUM("default")*100.0 / COUNT(*),2) AS default_rate

FROM loans
GROUP BY loan_size;
```

Default rates rise with loan size: **Small** loans default at **16.58%**, **Medium** loans at **19.78%**, and **Large** loans at **24.45%**. Larger exposures therefore carry disproportionately higher credit risk, which means concentration in big-ticket loans can significantly amplify portfolio losses if underwriting is not conservative enough for these segments.
